Respecte a la variable pes, després de consultar el manual de com han realitzat les enquestes, he pogut esbrinar el següent:

''El factor es defineix com la suma ponderada de les entrevistes del qüestionari més una estimació de les famílies que no van respondre al qüestionari, dividida per la suma ponderada de les entrevistes del qüestionari. L'estimació de les famílies que no hi van respondre es calcula multiplicant la proporció d'elegibilitat entre els casos coneguts per la suma ponderada de les famílies amb elegibilitat desconeguda.''



En resum, l'Screener Nonresponse Adjustment Factor és una manera d'ajustar els pesos de les famílies que van respondre al qüestionari per tenir en compte les famílies que no ho van fer, però al no tenir constància de algunes dades no puc ajustar la variable a un pes 'real".

Passem a fer el metodo del codo per veure quants clústers hauríem de fer servir al K-Means.

```python
def elbow_plot(df):
    sse = {}
    sse_r = []
    for k in range(1, 16):
        kmeans = KMeans(n_clusters=k, random_state=1)
        kmeans.fit(df)
        sse[k] = kmeans.inertia_
        sse_r.append(kmeans.inertia_)
    
    plt.title('Elbow plot for K selection')
    plt.xlabel('Número de clústers')
    plt.ylabel('SSE')
    sns.pointplot(x=list(sse.keys()),
                 y=list(sse.values()))
    plt.show()
    return sse_r
sse = elbow_plot(df_final)
```

```python
kl = KneeLocator(range(1, 16), sse, curve="convex", direction="decreasing")
kl.elbow
```

Fem K-means amb els 4 clusters que ens ha indicat.

```python
k_means = KMeans(n_clusters=4, random_state=60)
k_means.fit(df_final)
```

```python
centroides = k_means.cluster_centers_
y_pred = k_means.predict(df_final)
y_pred
```

```python
df['Cluster'] = y_pred
df.head()
```
